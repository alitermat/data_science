Lee el people, dejando de manera explícita el encoding y el separador.

```python
df_people = pd.read_csv("people.csv", encoding= 'utf-8', sep=';')
df_people
```

```python
df_people.columns
```

```python
df_people = df_people.drop('Unnamed: 0', axis=1)
```

```python
df_people.columns = df_people.columns.str.lower().str.replace(' ', '_')
df_people.columns
```

Saca la edad de las personas y guardalo en una columna llamada edad

```python
df_people.info()
```

```python
df_people['date_of_birth'] = pd.to_datetime(df_people['date_of_birth'], dayfirst=True)
```

```python
df_people.info()
```

```python
from datetime import datetime
current_date = datetime.now()
df_people['age'] = ((current_date - df_people['date_of_birth']).dt.days / 365.25).astype(int)
```

```python
df_people
```

Genera una columna que sea 1 si es mayor de edad y 0 si es menor

```python
df_people['is_adult'] = (df_people['age'] >= 18).astype(int)
df_people
```

Podrías calcular lo mismo pero con funciones lambda?

```python
df_people['is_adult'] = df_people['age'].apply(lambda x: 1 if x >= 18 else 0)
df_people
```

Lee el dataset organizations

```python
df_organizations.columns
```

```python
df_organizations = df_organizations.drop('Unnamed: 0', axis=1)
```

```python
df_organizations.columns = df_organizations.columns.str.lower().str.replace(' ', '_')
df_organizations.columns
```

Filtra el dataset de organizations seleccionando solo las columnas = ['Organization Id', 'Name'] y el dataset people seleccionando las columnas = ['User Id', 'First Name', 'Last Name', 'Sex', 'edad','Date of birth', 'is_mayor_edad', 'Id_organization'].

```python
df_organizations_filtered = df_organizations[['organization_id', 'name']]
df_organizations_filtered
```

```python
df_people_filtered = df_people[['user_id', 'first_name','last_name','sex','age','date_of_birth','is_adult','id_organization']]
df_people_filtered
```

Une mediante merge organizations con people. La idea es traer el nombre de la organización al dataset de people. Queremos quedarnos solo con los valores que son coincidentes en ambos datasets, es decir, cuando el valor que se utiliza en la columna de unión se encuentra en los dos datasets.

```python
df_people.head(1)
```

```python
df_organizations.head(1)
```

```python
df_organizations.rename(columns={'organization_id': 'id_organization'}, inplace=True)
```

```python
merged_data = pd.merge(df_people, df_organizations, on='id_organization', how='inner')
merged_data
```

Ahora, vamos a realizar la misma operación, pero en lugar de quedarnos solo con los coincidentes, vamos a quedarnos con todas las personas independientemente de si las encuentra en organizations o no.

```python
df_merged = pd.merge(df_people, df_organizations, on='id_organization', how='left')
df_merged
```

Media de edad de trabajadores por cada organización

```python
mean_age_by_organization = merged_data.groupby('name')['age'].mean().sort_values(ascending=False).round().astype(int).reset_index()
mean_age_by_organization
```

Que tipo de gráfico sería el mejor para plasmar esta información. Realízalo con seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
bar_plot = sns.barplot(x='name', y='age', data=mean_age_by_organization, errorbar=None)
bar_plot.set(xlabel='Nombre de la organización', ylabel='Media de edad')
plt.title('Media de edad de trabajadores por organización')
bar_plot.set_xticklabels(bar_plot.get_xticklabels(), rotation=45, horizontalalignment='right');
plt.show()
```
