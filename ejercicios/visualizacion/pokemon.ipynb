### Ejercicio 13:
- Calcula el porcentaje de victoria de cada pokemon y luego une esa información con el dataset de pokemons

Pistas: 
+ Calcula el número de victorias de cada pokemon
+ Calcula el número de apariciones de cada pokemon tanto en la primera como en la segunda columna 
+ Suma el total de apariciones de cada pokemon en las dos columnas, siendo esto el número total de combates
+ El porcentaje de victoria será el número de victorias entre el número de combates

```python
combat.head(2)
```

```python
#Calcula el número de victorias de cada pokémon.
wins_count = combat['Winner'].value_counts()
```

```python
# Calcula el número de apariciones en las columnas 'First_pokemon' y 'Second_pokemon'
first_pokemon_appearances = combat['First_pokemon'].value_counts()
second_pokemon_appearances = combat['Second_pokemon'].value_counts()
```

```python
#Suma el total de apariciones de cada pokémon en las dos columnas, siendo esto el número total de combates
total_appearances = first_pokemon_appearances.add(second_pokemon_appearances, fill_value=0)
```

```python
# Calcular el porcentaje de victoria de cada pokémon
win_percentage = wins_count / total_appearances*100
```

```python
combat['win_percentage']= combat['Winner'].map(win_percentage)
combat['win_percentage']= combat['win_percentage'].round(2)
combat
```

```python
df['win_percentage'] = combat['win_percentage']
df.head()
```
